```python
import pandas as pd
import numpy as np
import seaborn as sns
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("/content/movies.csv")
```

```python
selected_features = ["genres","keywords","original_title","cast","director"]
print(selected_features)
```

```python
for feature in selected_features:
  df[feature] = df[feature].fillna(" ")
```

```python
combine_feature = df["genres"]+" "+df["keywords"]+" "+df["original_title"]+" "+df["cast"]+" "+df["director"]
print(combine_feature)
```

```python
vectorizer = TfidfVectorizer()
feature_vectorizer = vectorizer.fit_transform(combine_feature)
print(feature_vectorizer)
```

Input Movie

```python
movie_name = input("Enter the movie name: ")
```

```python
all_movies = df["title"].tolist()
```

Find the close match

```python
find_close_match = difflib.get_close_matches(movie_name,all_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

Find Index based on movie name

```python
find_index_no = df[df.title == close_match]["index"].values[0]
print(find_index_no)
```

Getting list of the similar Movies

```python
similar_movies_score = cosine_similarity(feature_vectorizer)
print(similar_movies_score)
```

```python
get_similar_movies = list(enumerate(similar_movies_score[find_index_no]))
print(get_similar_movies)
```

```python
prioratize_score = sorted(get_similar_movies, key=lambda x:x[1], reverse=True)
print(prioratize_score)
```

```python
print("The recommend movies for you: \n")

i = 1
for movie in prioratize_score:
  index = movie[0]
  # Get the movie name based on title
  title_of_movie = df[df.index == index]["title"].values[0]

  if(i<30):
    print(i,".",title_of_movie)
    i+=1
```
